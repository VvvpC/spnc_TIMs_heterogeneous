### 库

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plot_style_config import set_pub_style, get_figsize
```

### 绘制varimp图像

```python
# 读取varimp.csv文件
varimp_df = pd.read_csv('varimp.csv')
varimp_df = varimp_df.sort_values('percentage', ascending=False)

# 绘制varimp图像
set_pub_style('double')
fig, ax = plt.subplots(figsize=get_figsize('double'))

# 使用柱状图来绘制
ax.barh(varimp_df['variable'], varimp_df['percentage'])
ax.invert_yaxis()   # 让最高的重要性显示在最上面

ax.set_xlabel('Percentage')
ax.set_ylabel('Variable')
plt.tight_layout()
plt.show()
```

```python
# 绘制出预测结果的图像
# 读取预测结果
predictions_df = pd.read_csv('pred_oof_train.csv')

# 绘制预测结果的图像
set_pub_style('double')
fig, ax = plt.subplots(figsize=get_figsize('double'))

ax.plot(np.linspace(0.0,5.0), np.linspace(0.0,5.0), 'k--', linewidth=2)
ax.plot(predictions_df['averagenrmse'], predictions_df['y_oof_pred'], 'o')

ax.set_xlabel('Target')
ax.set_ylabel('Predicted')
plt.tight_layout()
# 保存图像
fig.savefig('pred_oof_train.png', dpi=300)

plt.show()
```
